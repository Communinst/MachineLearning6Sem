3. Напишите функцию, которая принимает матрицу и производит нормировку её элементов (т.е. $A'_{ij} = A_{ij} / \sum_{i} \sum_{j} A_{ij}$)

```python
import numpy as np

def normalizeMatrix (sample) :
  #volume
  if not isinstance(sample, np.ndarray) :
      raise TypeError("Input data's to be numpy array type.")
  if sample.ndim != 2 :
      raise ValueError("Only 2D arrays.")

  matrixSum = np.sum(sample)
  if (matrixSum == 0) :
    ValueError("Matrix sum's to be non zero value.")

  normalized = sample / matrixSum
  return normalized


np.random.seed(48)
sample = np.random.rand(3, 3) * 100
#print("Initial one:\n", sample)
print("Eventual one:\n", normalizeMatrix(sample))
```

4. Напишите функцию, которая принимает 3 трехмерных вектора $x, y, z$ и вычисляет объем пирамиды, образованной этими векторами (используйте смешанное произведение векторов)

```python
import numpy as np

def tetrahedron_volume(x, y, z) :
  if x.shape != (3,) or y.shape != (3,) or z.shape != (3,):
    raise ValueError("Vectors are to be 3d")

  mixedMult = np.dot(x, np.cross(y, z))
  return 1 / 6 * mixedMult

x = np.array([1, 0, 0])
y = np.array([0, 1, 0])
z = np.array([0, 0, 1])

print(f"Tetrahedron volume: {tetrahedron_volume(x, y, z)}")
```

10. Даны 2 массива Х и Y, постройте матрицу Коши $C$ ($C_{ij} = \frac{1}{x_i - y_j}$)

```python
import numpy as np

def buildCauchyMatrix(x, y) :

  # broadcasting is the most effective up to 1e(?)
  # rotate X and diff matrix
  deff = x[:, np.newaxis] - y

  with np.errstate(divide = 'ignore', invalid = 'ignore') :
    C = 1 / deff
    C[np.isinf(C)] = 0

  return C

np.random.seed(48)
x = np.random.randint(0, 6, size = 5)
y = np.random.randint(0, 6, size = 6)
print("X: ", x, "\nY: ", y)

print(buildCauchyMatrix(x, y))
```

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

```python
countries = pd.read_csv('countries.csv')
covid = pd.read_csv('covid-country-data.csv')
```

1. Сделайте круговую диаграмму, на которой были бы отражено число заболеваний COVID-19 по континентам

```python
import pandas as pd
import matplotlib.pyplot as plt

mergedData = pd.merge(countries, covid, on='location')
continentCases = mergedData.groupby('continent')['total_cases'].sum()

plt.figure(figsize=(10, 8))
continentCases.plot.pie(autopct='%1.1f%%', startangle=140, fontsize=12, ylabel='')
plt.title('COVID-19 Cases by Continent')
plt.show()
```

2. Постройте диаграмму, на которой в порядке убывания были бы отмечены число тестов на COVID-19 в России, Китае, США, Индии и Нигерии

```python

countriesOfInterest = ['Russia', 'China', 'United States', 'India', 'Nigeria']
filteredData = covid[covid['location'].isin(countriesOfInterest)]

sortedData = filteredData.sort_values(by='total_tests', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(sortedData['location'], sortedData['total_tests'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Total Tests')
plt.title('Total COVID-19 Tests by Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

3. Визуализируйте зависимость ВВП страны и случаев заболевания COVID-19

```python

mergedDataGdp = pd.merge(countries, covid, on='location')
mergedDataGdp = mergedDataGdp.dropna(subset=['gdp_per_capita', 'total_cases'])

plt.figure(figsize=(12, 8))
plt.scatter(mergedDataGdp['gdp_per_capita'], mergedDataGdp['total_cases'], alpha=0.5, color='green')
plt.xlabel('GDP per Capita')
plt.ylabel('Total COVID-19 Cases')
plt.title('Relationship between GDP per Capita and Total COVID-19 Cases')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which='both', ls='--')
plt.show()
```

4. Постройте гистограмму распределения количества смертей на один миллион человек

```python

mergedData = pd.merge(countries, covid, on='location')
mergedData['deaths_per_million'] = (mergedData['total_deaths'] / mergedData['population']) * 1e6

plt.figure(figsize=(12, 8))
plt.hist(mergedData['deaths_per_million'].dropna(), bins=30, color='purple', alpha=0.7, edgecolor='black')
plt.xlabel('Deaths per Million')
plt.ylabel('Frequency')
plt.title('Distribution of COVID-19 Deaths per Million')
plt.grid(True, which='both', ls='--')
plt.show()
```

6. Сравните распределения кол-ва смертей на 1000 человек в Европе и Африке

```python
mergedData = pd.merge(countries, covid, on='location')
mergedData['deaths_per_thousand'] = (mergedData['total_deaths'] / mergedData['population']) * 1e3

europe_deaths = mergedData[mergedData['continent'] == 'Europe']['deaths_per_thousand']
africa_deaths = mergedData[mergedData['continent'] == 'Africa']['deaths_per_thousand']

plt.figure(figsize=(14, 7))

#EU
plt.subplot(1, 2, 1)
plt.hist(europe_deaths.dropna(), bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('Deaths per 1,000 People')
plt.ylabel('Frequency')
plt.title('Distribution of COVID-19 Deaths per 1,000 People in Europe')
plt.grid(True, which='both', ls='--')

#AF
plt.subplot(1, 2, 2)
plt.hist(africa_deaths.dropna(), bins=20, color='green', alpha=0.7, edgecolor='black')
plt.xlabel('Deaths per 1,000 People')
plt.ylabel('Frequency')
plt.title('Distribution of COVID-19 Deaths per 1,000 People in Africa')
plt.grid(True, which='both', ls='--')

plt.tight_layout()
plt.show()
```
